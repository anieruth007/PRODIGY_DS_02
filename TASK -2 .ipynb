```python
import pandas as pd
df = pd.read_csv("alcohol.csv")
print(df)
```

```python
df.head()
```

```python
df1 = pd.DataFrame(df)
df1
```

```python
year_data = df.loc[df1['Year'] == 2015]
print(year_data)
```

```python
country_data = df[df1['Entity'] == 'India']
print(country_data)
```

```python
df1.rename(columns={"Total alcohol consumption per capita (liters of pure alcohol, projected estimates, 15+ years of age)" : "consumption"}, inplace=True)
print(df1)
```

```python
highest_consumption = df1['consumption'].max()
highest_country = df1[df1['consumption'] == highest_consumption]['Entity'].iloc[0]
highest_year = df1[df1['consumption'] == highest_consumption]['Year'].iloc[0]

print(f"Highest Consumption: {highest_consumption}")
print(f"Country: {highest_country}")
print(f"Year: {highest_year}")
```

```python
df1.tail(10)
```

```python
df1['consumption'] = pd.to_numeric(df1['consumption'], errors='coerce')
mean_consumption = df1['consumption'].dropna().mean()
print(f"Mean Consumption: {mean_consumption}")
```
